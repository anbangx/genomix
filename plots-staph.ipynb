```python
import scipy
import re
import string
```

```python
comptable = string.maketrans('acgtACGT', 'TGCATGCA')
revcomp = lambda seq: seq.translate(comptable)[::-1]
genome = open('/Users/jakebiesinger/in-staph/genome.fasta').read().replace('\n', '')
genome = genome + '\n' + revcomp(genome)
```

```python
d = []
with open('/Users/jakebiesinger/out-staph/build-merge-tip-bubble2/FINAL-06-MERGE/data') as infile:
    for line in infile:
        match = re.search('.*kmer:([ACGT]*).*cov:([\d.]+)', line)
        kmer, coverage = match.groups()
        readcount = line.count('readSeq')
        d.append((len(kmer), float(coverage), kmer in genome, genome.count(kmer), readcount))
print len(d)
d = scipy.array(d)
print d.shape
```

```python
ingenome = d[d[:,2] == 1]
outgenome = d[d[:,2] == 0]

figure(dpi=600)
scatter(ingenome[:,0], ingenome[:,1], c='r', alpha=.1)
scatter(outgenome[:,0], outgenome[:,1], c='b', alpha=.1)
xlabel('Kmer size')
ylabel('coverage')

figure(dpi=600)
scatter(ingenome[:,0], ingenome[:,1], c='r', alpha=.1)
ylim(0, 400)
xlim(21, 100)
xlabel('Kmer size')
ylabel('coverage')

figure(dpi=600)
scatter(outgenome[:,0], outgenome[:,1], c='b', alpha=.1)
ylim(0, 400)
xlim(21, 100)
xlabel('Kmer size')
ylabel('coverage')

figure(dpi=600)
scatter(ingenome[:,0], ingenome[:,1], c='r', alpha=.1)
scatter(outgenome[:,0], outgenome[:,1], c='b', alpha=.1)
ylim(0, 400)
xlim(21, 100)
xlabel('merged length')
ylabel('coverage')
title('red=ingenome, n=' + str(len(ingenome)) + ' blue=outgenome, n=' + str(len(outgenome)))
```
